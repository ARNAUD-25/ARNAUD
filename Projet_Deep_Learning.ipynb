# Result_collect

## Ici nous allons predire Result_collect

#### Nous prenons comme x_train notre data qui enleve les colonnes et les lignes de Result_collect n'ayant pas des valeurs manquantes avec dropna

```python
x_train = data.dropna(subset=['Result_collect'])
x_train
```

#### Et prenons comme y_train la colonne de Result_collect n'ayant pas des valeurs manquantes tout en transformant notre variable en entier

#### Ici nous prenons x_test qui va nous permettre de predire nos notre cible Result_collect et x_test est obtenu en prenant les lignes et les colonnes de notre data qui ont des valeurs manquantes sur la colonne de Result_collect. b nous a permis a recupere les lignes des valeurs manquantes Result_collect et x_test est obtenu en prenant notre b moins toutes les variables que je cherche a predire.

```python
b=data.loc[data['Result_collect'].isnull()]
x_test=b.drop(['Result_collect', 'Dest_principal','implicaion_femme'], axis=1)
x_test
```

#### Nous prenons maintenant notre x_train ci-haut et nous enlevons nos 3 variables a predire.  

```python
x_train = x_train.drop(['Result_collect', 'Dest_principal','implicaion_femme'], axis=1)
x_train
```

```python
x_train.shape
```

#### A cette etape, nous cherchons a categoriser notre y_train, par categoriser nous convertissons le vecteur de classe y_train (entiers) en matrice de classe binaire avec to_categorical.

```python
y_train=to_categorical(y_train)
y_train
```

```python
y_train.shape
```

#### Enfin nous utilisons maintenant notre reseau de neurone.  Nous utilisons 3 couches dont la couche d'entree a 29 neurones, la couche cachee a 19 et la couche de sortie a 9 neurones correspondant a la colonne de y_train quand on categorise (nombre total de classe).

```python
classifier=Sequential()
classifier.add(Dense(29, activation='relu', input_shape=(11,)))
classifier.add(Dense(19, activation='relu', input_shape=(11,))) # essayer avec la fonction leakyrelu
classifier.add(Dense(9, activation='softmax')) # essayer avec la fonction softmax
```

```python
classifier.compile(loss="categorical_crossentropy",optimizer="adam",metrics=['accuracy'])
#classifier.compile(optimizers="adam",optimizer="binary_crossentropy",metrics=['accuracy'])
```

#### Nous entrainons enfin notre modele avec la commande fit

```python
classifier.fit(x_train, y_train, epochs=9, batch_size=5)
```

```python
accuracy=classifier.evaluate(x_train,y_train)
```

#### Nous obtenons un accuracy de 99,96%, qui montre que notre modele a tres bien appris et que les poids de connexions trouves sont les bons

```python
print("\n%s: %.2f%%" %(classifier.metrics_names[1],accuracy[1]*100))
```

#### Nous procedons enfin a la prediction avec notre x_test. 

```python
y_pred=classifier.predict(x_test)
y_pred=classifier.predict_classes(x_test)
```
