```python
df_= Copie_Afrique[['superficie','region']].groupby(['region']).min()
n=Copie_Afrique.reset_index(level = ['pays'],inplace =False) 
nn=n[["pays",'superficie','region']].merge(df_)
nn
```

### Calculate and present in the same Dataframe: The average of the population (µ) of each region, the standard deviation (σ) and the calculation of the coefficient of variation (Cv).

###  Calculer et présenter dans un même Dataframe : La moyenne de la population (µ) de chaque région, l’ecart type (σ) et le calcul du coefficient de variation (Cv). 

```python
#cv =  lambda x: np.std(x) / np.mean(x)
df = Copie_Afrique[['population','region']].groupby(['region']).agg([np.mean, np.std])
df.columns = ['moyenne', 'ecart type']
print("\n")
print("                                 population                                     ")

df
```

```python
df['coef_var']=df['ecart type']/df['moyenne']
df
```

```python
df3=pd.merge(nn, df, on='region', how='right')
print("\n\n                min                              population ")


df3
```

## Conclusion

#### Starting from coefficient of variation; we find that the populations of North Africa are not very dispersed around the average and those of West Africa are very dispersed around the average. / Partant de coefficient de variation ; nous constatons que les populations de l'Afrique du nord sont peu dispersées autour de la moyenne et celles de l'Afrique de l'ouest sont très dispersées autour de la moyenne.
#### The measurement of the standard deviation is an index of dispersion sensitive to the average value and to the unit of the values used, but the coefficient of variation allowed us to obtain a general index independent of the units of measurement used. / La mesure de l'ecart type est un indice de dispersion  sensible à la valeur moyenne et à l'unité des valeurs utilisées mais le coeficient de variation nous a permi d'obtenir un indice général independant des unités de mésure employées.
